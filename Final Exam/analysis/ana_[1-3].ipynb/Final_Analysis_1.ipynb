## Impact of Weather on Renewable or NON Renewable Energy ?

### Question - Is there a realtion between Weather & Stock Market?

#### Answer - 
- The traditional efficient markets view says no, with minor qualifications. 
- If sunlight affects the weather, it can affect agricultural and perhaps other weather-related firms. 
- But in modern economies in which agriculture plays a modest role, it seems unlikely that whether it is cloudy outside the stock exchange today should affect the rational price of the nationâ€™s stock market index.


### We will analysize the relationship between Weather & Stock Market. If exist,  Then which sectors affects the most

#### Here we considering 3 major Sectors.
- CRUDE OIL (Non Renewable)

```python
import os
import pandas as pd
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
from plotly.graph_objs import Scatter, Figure, Layout
from plotly.graph_objs import *
init_notebook_mode(connected=True)
import plotly.graph_objs as go
import warnings
warnings.filterwarnings('ignore')
```

```python
# Getting Current Working Directory
cwd =  os.getcwd()

# Reading my weather File which is stored in cwd + final / WeatherData
df = pd.read_csv(cwd + "//final//data//weatherData//weather.csv", parse_dates=['Date'], usecols=[0,1,2,4])

# This how data looks like.
df.head(1)
```

### Plotting Max Temp for all years (2012-2017)
 - In order to find the max variation in particular year

```python
# Now task is to Group The Dataframe on basis of month
df['dayMonth'] = df['Date'].dt.strftime('%b').astype(str).str.cat(df['Date'].dt.day.astype(str), sep='-')
df.head(1)
```

```python
# Creating Data Sets for Scatter Plot
y2012 = go.Scatter(
    x=df.dayMonth,
    y=df[df.Date.dt.year == 2012]['Max Temp'],
    name = "2012 Max Temp",
    line = dict(color = 'Blue'),
    opacity = 0.8)

y2014 = go.Scatter(
    x=df.dayMonth,
    y=df[df.Date.dt.year == 2014]['Max Temp'],
    name = "2014 Max Temp",
    line = dict(color = 'Green'),
    opacity = 0.8)

y2016 = go.Scatter(
    x=df.dayMonth,
    y=df[df.Date.dt.year == 2016]['Max Temp'],
    name = "2016 Max Temp",
    line = dict(color = 'Red'),
    opacity = 0.8)

layoutMaxTemp = go.Layout(
    title='Per Day Max Tempearture in Degree (C) [2012, 2014, 2016]',
)

dataPerYear = [y2012, y2014, y2016]
maxTemPerYear = go.Figure(data = dataPerYear, layout=layoutMaxTemp)
#image.save_as(stockFig, 'Per Day Max Temp.png')
iplot(maxTemPerYear)
```

### After Analyzing the above graph:
- Every second year tempeature was increasing.
- 4(2012)
- 7(2014)
- 19(2016)

#### As in above graph on per day Plotting difference is very minute which is not clearly visible.
#### So in order to make it visible I will caluclate mean Tempearture for Every Year.



```python
# Calculating Mean for each year in order to plot
dfMeanYear = df.groupby([df.Date.dt.year]).mean()
dfMeanYear = dfMeanYear.reset_index()
dataMean = [go.Bar(
            x=dfMeanYear['Date'],
            y=dfMeanYear['Max Temp']
    )]

layoutMean = go.Layout(
    title='Annual Mean Tempearture in Degree (C)',
)

fig = go.Figure(data = dataMean, layout=layoutMean)
#image.save_as(fig, 'Annual Mean Temp.png')
iplot(fig, filename='basic-bar')
```

### As we can see now Mean Temp is Increasing from 2012 to 2016.

#### On the basis of above observation I will plot Stock VS Weather for Year (2102, 2016)

```python
# Reading the CVX (OIL) data & merging with weather w.r.t Date.
df2 = pd.read_csv(cwd + "//final//data//stockData//CVX.csv", parse_dates=['Date'])

# now will merger two DataSet into one Data Frame (Weather & Stock)
df3 = pd.merge(df, df2, on='Date', how='outer')

df3.head(1)

# As I can see lot of NAN Values I will drop all NAN

df3 = df3.dropna()
df3.head(1)
```

```python
# Creating Data sets for Scatter Plot.
weather2012 = go.Scatter(
    x=df3['dayMonth'],
    y=df3[df3.Date.dt.year == 2012]['Max Temp'],
    name = "2012 Max Temp",
    line = dict(color = 'Green'),
    opacity = 0.8
)

weather2016 = go.Scatter(
    x=df3['dayMonth'],
    y=df3[df3.Date.dt.year == 2016]['Max Temp'],
    name = "2016 Max Temp",
    line = dict(color = 'Red'),
    opacity = 0.8
)

stock2012 = go.Scatter(
    x=df3['dayMonth'],
    y=df3[df3.Date.dt.year == 2012]['Close'],
    name = "2012 Stock CVX(OIL)",
    line = dict(color = 'Green'),
    opacity = 0.8
)

stock2016 = go.Scatter(
    x=df3['dayMonth'],
    y=df3[df3.Date.dt.year == 2016]['Close'],
    name = "2016 Stock CVX(OIL)",
    line = dict(color = 'Red'),
    opacity = 0.8
)

layout = go.Layout(
    title='Weather VS OIL Stock',
    yaxis=dict(
        title='Stock of CVX(OIL) in $'
    ),
    yaxis2=dict(
        title='Weather of NYC in (C)',
        titlefont=dict(
            color='rgb(148, 103, 189)'
        ),
        tickfont=dict(
            color='rgb(148, 103, 189)'
        ),
        overlaying='y',
        side='right'
    )
)
stockData = [weather2012, weather2016, stock2012, stock2016]
stockFig = go.Figure(data=stockData, layout=layout)
#py.image.save_as(stockFig, 'Weather VS OIL Stock.png')
iplot(stockFig)
```
