- Under package Gutenberg, see what files are in it.
- Find out how many words and how many unique words in file 'austen-persuasion.txt'.
- Calculate the frequency for each words and show the top 10 most frequent words and their frequency.
- Keep the punctuation and stop words in this question.



```python
import nltk #first import nltk
gutenberg = nltk.corpus.gutenberg # import package gutenberg
gutenberg.fileids() # get all files from Gutenberg library
```

```python
Words = gutenberg.words("austen-persuasion.txt") # Getting all words from austen-persuasion.txt
count = len(Words)
print(count)
```

```python
unique_words = {} # I created a Dictionary to store all unique words.
for word in Words: # Looped over all Words to find the unique one's and added into the Dictionary.
    if word in unique_words:
        unique_words[word] += 1
    else:
        unique_words[word] = 1
countUnique = len(unique_words)
print('Count of Unique words using for Loop is', countUnique)

# This can also be achieve by using Set function which is Unordered collections of unique elements.
countWords = set(Words)
uniqueCount = len(countWords)
print('Count of unique words using Set Function is', uniqueCount)
```

```python
print(unique_words) # Variable unique_words contains the frequency of each word in the document sten-persuasion.txt
```

```python
sortedDict = sorted(unique_words.items(), key=lambda x: (-x[1], x[0])) # sorted is used to sort dictionary in descending order.
print(sortedDict)
```
