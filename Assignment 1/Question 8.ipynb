- Use nps_chat and wordlist corpus.
- If you find a word in nps_chat but not in wordlist, it is considered an unusual word. (e.g. Heyyyy, gr8, 4ever can be considered unusual words for this analysis)
- Find out how many unusual words people speak in chat (notice: usernames should not be counted)
- Bonus: What if I want to count : ) , : ( all these emoticons in. (optional)

```python
import nltk #import nltk
nps_chat = nltk.corpus.nps_chat #import nps_chat
wordlist = nltk.corpus.words # import wordlist



uniqueWordlist = set([word.lower() for word in list(wordlist.words())]) # Unique word list
uniqueNpsChat = set([chat.lower() for chat in list(nps_chat.words())]) # unique chatlist

unusualWord = [unWord for unWord in uniqueNpsChat if unWord not in uniqueWordlist] # unusualWord present in chat not in wordlist

print(unusualWord)
```

```python
# Defining function to remove all users and special characters
import re
def unWord(elem):
    pattern=re.compile('\W+|u(\d+)|(\d+)|\\\w+')
    return  re.match(pattern,elem)

noUser = [user for user in unusualWord if not unWord(user)]
print(noUser) # All unusualWord with no user and punctuation
```
